```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import seaborn as sns
import sklearn
import numpy as np
```

```python
sea_lvlDF = pd.read_csv("CSIRO_Recons_gmsl_yr_2011.csv")
sea_lvlDF.columns =["Year","Sea Lvl","uncertainty"] 
sea_lvlDF.head()
```

```python
sns.regplot(x='Year', y='Sea Lvl', data=sea_lvlDF, fit_reg=False)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = np.asarray(train_test_split(sea_lvlDF['Year'], sea_lvlDF['Sea Lvl'], test_size=0.1))

reg = LinearRegression()
reg.fit(X_train.values.reshape(-1,1), y_train.values.reshape(-1,1))

```

```python
sns.regplot(x=sea_lvlDF['Year'], y=sea_lvlDF['Sea Lvl'], data=sea_lvlDF, fit_reg=False)
plt.plot(np.arange(1865,2020).reshape(-1,1), reg.predict(np.arange(1865,2020).reshape(-1,1)))
plt.show()
```

```python
c_eDF = pd.read_csv("global.1751_2014_modified.csv")
c_eDF.columns =["Year","carbon emissions"] 
c_eDF = c_eDF.loc[c_eDF['Year'] >= 1800]
c_eDF.head()
```

```python
sns.regplot(x='Year', y='carbon emissions', data=c_eDF, fit_reg=False)
plt.show()
```

```python
X_train, X_test, y_train, y_test = np.asarray(train_test_split(c_eDF['Year'], c_eDF['carbon emissions'], test_size=0.1))

reg = LinearRegression()
reg.fit(X_train.values.reshape(-1,1), y_train.values.reshape(-1,1))

```

```python
sns.regplot(x=c_eDF['Year'], y=c_eDF['carbon emissions'], data=c_eDF, fit_reg=False)
plt.plot(np.arange(1800,2020).reshape(-1,1), reg.predict(np.arange(1800,2020).reshape(-1,1)))
plt.show()
```
