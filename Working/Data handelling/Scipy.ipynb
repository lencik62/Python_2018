<h1>Multivaraite norm</h1>

```python
from scipy.stats import multivariate_normal as mvn
```

```python
cov =  np.array([[1,0.8],[.08,3]])
```

```python
mu =  np.array([0,2])
```

```python
m_r = mvn.rvs(mean = mu, cov = cov, size  = 1000)
plt.scatter(m_r[:,0],m_r[:,1])
plt.axis('equal')
plt.show()
```
