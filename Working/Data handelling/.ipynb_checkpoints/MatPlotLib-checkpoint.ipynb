```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_s  =  np.linspace(0,10,10)
more_x_s = np.linspace(0,10,100)

y =  np.sin(x_s)
better_y = np.sin(more_x_s)

plt.xlabel("Time")
plt.ylabel("F(x)")
plt.title("Sin(x)")
plt.plot(x_s,y)
plt.plot(more_x_s,better_y)
plt.show()
```

```python
plt.clf()
x = np.arange(10)

plt.gca().set_color_cycle(['red', 'green', 'blue', 'yellow'])

plt.plot(x, x)
plt.scatter(x,x)
plt.plot(x, 2 * x)
plt.plot(x, 3 * x)
plt.plot(x, 4 * x)

plt.legend(['y = x', 'y = 2x', 'y = 3x', 'y = 4x'], loc='upper left')

plt.show()
```
